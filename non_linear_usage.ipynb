```python
from scipy.optimize import minimize
import numpy as np
```

```python
x0 = np.zeros(2)
x0
```

```python
p = 100
def obj(x):
    return -x[0]*x[1]
def con(x):
    return x[0]+x[1]-p  
cons = {'type': 'eq', 'fun': con}
```

```python
res = minimize(obj, x0, method='SLSQP', constraints=[cons], options={'maxiter': 100, 'ftol': 1e-06, 'iprint': 1, 'disp': True, 'eps': 1.4901161193847656e-08}) 
```
